```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200, 56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.xlabel("ya")
plt.ylabel("maa")
plt.plot(age,salary);
```

```python
fig, ax = plt.subplots()
ax.plot(age,salary,"r");
ax.set_xlabel("maa")
ax.set_ylabel("maalar")
ax.set_title("Yalara gre maalar");
```

```python
fig, ax = plt.subplots()
ax.plot(age,salary, 'r')
ax.set_xlabel("age")
ax.set_ylabel('salary')
ax.set_title("Salary by Age")
plt.show()
```

```python
x = np.arange(0,11)
y = x.copy()
z = x*2
t = np.log(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
ax.plot(x,t)
ax.set_xlabel("X")
ax.set_ylabel("Y or Z")
ax.set_title("Title")
plt.show()
```

```python
fig = plt.figure()      # ax = fig.add_axes([left, bottom, widht, height])
                                          # left, bottom, width, height (range 0 to 1)
ax1 = fig.add_axes([0.0, 0.0, 1, 1])
ax2 = fig.add_axes([1.1, 0.3, 0.4, 0.4])
ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")
ax2.plot(age,salary, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0, 0.0, 1, 1])
ax2 = fig.add_axes([0.5, 0.2, 0.4, 0.3])
# 2. satrda ki 0.5 saa-sola pozisyonlar, 0.2 yukar-aaya pozisyonlar, 0.4 genilik, 0,3 ykseklikleri. 
ax1.plot(age,salary, "r")
ax1.set_xlabel("Age")
ax1.set_ylabel("Salary")
ax1.set_title("Salary by Age")
ax2.plot(age,salary, "b");
ax2.set_xlabel("Age")
ax2.set_ylabel("Salary_2")
ax2.set_title("Salary_2 by Age");
plt.show()
```

```python
# funcational olarak 2 satr 1 sutunluk :
plt.subplot(2,1,1) # 2 satr 1 sutun 1. satr
plt.plot(age,salary,"r")

plt.subplot(2,1,2) # 2 satr 1 sutun 2. satr
plt.plot(age,salary,"b")
plt.show()
```

```python
# OOP 
fig, ax = plt.subplots(2,1) # 2 satr 1 sutun
ax[0].plot(age,salary,"y")
ax[1].plot(age,salary,"b")
plt.show()
```

```python
fig, ax = plt.subplots(1,2) # 1 satr 2 sutun
ax[0].plot(age,salary,"g")
ax[1].plot(age,salary,"b")
plt.tight_layout(pad=5) # arasnda ki mesafeyi ayarlar. pad aralarndaki mesafeyi aar.
plt.show()
```

```python
fig , ax = plt.subplots(2,2) # 2 satr 2 sutun
ax[0][0].plot(age,salary,"r") # ilk satrn ilk sutunu
ax[1][1].plot(age,salary,"y") # 2. ci satrn 3. sutunu
plt.tight_layout(pad=5 , h_pad=5)
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
for x,y in axes:
    x.plot(age,salary, 'r')
    y.plot(age,salary, 'b')
    plt.tight_layout(pad=5 , h_pad=2.1)
```

```python
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(age,salary,"r")
plt.show()
```

```python
x = np.arange(10)
x
```

```python
# stil dzenleme 
fig, ax = plt.subplots(figsize=(12,6))
fig.suptitle("To change the line width, we can use the `linewidth` or `lw` keyword argument",fontsize=18)
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="red", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="red", lw=10)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x-1, color="green", lw=3, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.') # dash and dot
ax.plot(x, x-3, color="green", lw=3, ls=':') # dots
ax.plot(x, x-4, color="green", lw=3, ls='--') # dashes
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, x-1,marker='+',markersize=20)
ax.plot(x, x-2,marker='o',ms=20) #ms can be used for markersize
ax.plot(x, x-3,marker='s',ms=20,lw=0) # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=20)
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.suptitle("LineStyles & Markers & Markersize & Color",fontsize=16, y = 1.2)
plt.subplot(2,2,1)
plt.plot(age, salary, "r", ls="-", marker="+" ,markersize=30)
plt.subplot(2,2,2)
plt.plot(age, salary, "b",ls="-.", marker="o" ,markersize=30)
plt.subplot(2,2,3)
plt.plot(age, salary, "g",ls=":", marker="x" ,markersize=30)
plt.subplot(2,2,4)
plt.plot(age, salary, "y",ls="--" ,marker="*" ,markersize=30)
plt.tight_layout();
plt.show()
```

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].plot(x,x**2,x,x**3,x,x**4, 'r')
ax[1].set_xlim([1,3])
ax[1].set_ylim([1,50])
ax[0].axvline(x=5,ls="--")
ax[1].axhline(y=15,ls="--")
plt.show()
```

```python
plt.figure()
a1 = plt.subplot2grid((3,3),(0,0),colspan=2)
a2 = plt.subplot2grid((3,3),(0,2),rowspan=3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan =2, colspan=2)
x = np.arange(1, 10)
a1.plot(x, np.exp(x))
a1.set_title("exp")
a2.plot(x, x**2)
a2.set_title("square")
a3.plot(x, np.log(x))
a3.set_title("log")
plt.tight_layout()
plt.show()
```
